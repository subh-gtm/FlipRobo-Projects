```python
import nltk
import pandas as pd
import string
#!pip install gensim

from gensim.models import Word2Vec
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
%matplotlib inline
import re
two=open('two_cities.txt',encoding="utf-8").read()
```

```python
text = re.sub(r'\[[0-9]*\]',' ',two)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
sentences = nltk.sent_tokenize(two)
sentences
```

```python
sent_word = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
sent_word
```

```python
for i in range(len(sent_word)):
    sent_word[i] = [word for word in sent_word[i]
                    if word not in stopwords.words('english')]
```

```python
sent_word
```

```python
model = Word2Vec(sent_word, min_count=1)
words = model.wv.vocab
```

```python
from nltk.corpus import wordnet
synonyms = []
antonyms = []

for syn in wordnet.synsets("business"):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print('Synonyms =>',set(synonyms))
print('Antonyms =>',set(antonyms))
```

```python
from sklearn.manifold import  TSNE
vocab = ['work','without','people','might','always','among','place','prison','ever','put','way','face','even','men','sat','back','many','light','could','door']
def tsne_plot(model):
    labels = []
    wordvecs = []

    for word in vocab:
        wordvecs.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=3, n_components=3, init='pca', random_state=42)
    coordinates = tsne_model.fit_transform(wordvecs)

    x = []
    y = []
    for value in coordinates:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(15,8)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(2, 2),
                     textcoords='offset points', ha='right')

    plt.show()

tsne_plot(model)
```
